```
accounts = {}
print("welcome to bank")
print("1.Create Account")
print("2.Deposit")
print("3.Withdraw")
print("4.Check Balance")

choice = input("Enter choice: ")

if choice == "1":
    acc = input("Enter account number: ")
    pin = input("Set password: ")
    accounts[acc] = {"pin": pin, "bal": 0}
    print("Account created")

elif choice == "2":
    acc = input("Enter account number: ")
    amt = int(input("Enter amount: "))
    accounts[acc]["bal"] += amt
    print("Money deposited")

elif choice == "3":
    acc = input("Enter account number: ")
    amt = int(input("Enter amount: "))
    if amt <= accounts[acc]["bal"]:
        accounts[acc]["bal"] -= amt
        print("Money withdrawn")
    else:
        print("Insufficient balance")

elif choice == "4":
    acc = input("Enter account number: ")
    print("Balance:", accounts[acc]["bal"])

else:
    print("Invalid choice")
```

```
trans = [
    {"customer": "A", "amount": 300,  "status": "success"},
    {"customer": "B", "amount": 700,  "status": "success"},
    {"customer": "C", "amount": -300, "status": "failed"},
    {"customer": "D", "amount": 1300, "status": "success"}
]

clean_data = list(
    filter(lambda x: x["status"] == "success" and x["amount"] > 0, trans)
)

print("Clean Data:")
print(clean_data)
```
