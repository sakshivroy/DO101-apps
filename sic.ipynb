```
accounts = {}
print("welcome to bank")
print("1.Create Account")
print("2.Deposit")
print("3.Withdraw")
print("4.Check Balance")

choice = input("Enter choice: ")

if choice == "1":
    acc = input("Enter account number: ")
    pin = input("Set password: ")
    accounts[acc] = {"pin": pin, "bal": 0}
    print("Account created")

elif choice == "2":
    acc = input("Enter account number: ")
    amt = int(input("Enter amount: "))
    accounts[acc]["bal"] += amt
    print("Money deposited")

elif choice == "3":
    acc = input("Enter account number: ")
    amt = int(input("Enter amount: "))
    if amt <= accounts[acc]["bal"]:
        accounts[acc]["bal"] -= amt
        print("Money withdrawn")
    else:
        print("Insufficient balance")

elif choice == "4":
    acc = input("Enter account number: ")
    print("Balance:", accounts[acc]["bal"])

else:
    print("Invalid choice")
```

```
trans = [
    {"customer": "A", "amount": 300,  "status": "success"},
    {"customer": "B", "amount": 700,  "status": "success"},
    {"customer": "C", "amount": -300, "status": "failed"},
    {"customer": "D", "amount": 1300, "status": "success"}
]

clean_data = list(
    filter(lambda x: x["status"] == "success" and x["amount"] > 0, trans)
)

print("Clean Data:")
print(clean_data)
```

map() is used to apply a function to every item in a list.

Simple meaning:

map = do the same work on all elements

```
#transform data
#to create a new column tax(10%)

transform_data = list(

    map(lambda x: {**x, "tax": x["amount"] * 0.1}, clean_data)
)
print("Transformed Data:")
print(transform_data)
```

#.analyse

```
#add category 'high' ,'low'

category = list (map(lambda x: {**x, "category": "high" if x["amount"] > 300 else "low"}, transform_data))

print(category)
```

```
sorted_data = sorted(category, key=lambda x: x["amount"], reverse=True)

print("Sorted Data:")
print(sorted_data)
```

Local & Global

```
xv = 10   # global variable

def show():
    global xv
    xv = 5    # local variable (same name)
    print(xv)

show()
print(xv)
```

Closure function in py

```
def logger(level):
  def log(message):
    print(f"{level}: {message}")
  return log

log_info = logger("INFO")
log_error = logger("ERROR")

log_info("This is an info message")
log_error("This is an error message")
```
